```python
import pandas as pd
import os
import numpy as np
import datetime as dt
from filter_foia import filter_foia


cwd = os.getcwd()
cwd
```

```python
agency = 'Department of Veterans Affairs'
entity = 'VA'

processed_received = pd.read_csv("2024 all federal agencies foia-received-processed-and-pending-foia-requests.csv")
keep_cols = ['Fiscal Year', 'Number of Requests Received in Fiscal Year', 'Number of Requests Processed in Fiscal Year']
va_vol = filter_foia(processed_received, agency, keep_cols, entity)
va_vol.loc[len(va_vol)] = [dt.datetime(2025,1,1), 123789, 122552] #from 2025 Annual Report
va_vol
```

```python
va_vol = va_vol.sort_values('Fiscal Year', ascending=True)
va_vol['Increase Requests Processed'] = va_vol['VA Number of Requests Processed in Fiscal Year'].diff()
va_vol
```

```python
score_card_folder = r"C:\Users\OITNYNWilsoS\OneDrive - Department of Veterans Affairs\CRR Data Analytics (FOIA) - eFOIA\Score Card Power BI dashboards/"

with pd.ExcelWriter(
    "va volume.xlsx",
    engine="xlsxwriter",
    date_format="YYYY-MM-DD",
    datetime_format="YYYY-MM-DD"
) as writer:
    va_vol.to_excel(os.path.join(score_card_folder, "va volume.xlsx"), sheet_name="VA Volume", index=False)
```
